```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
# Define the input from the diagram
X = np.array([[1, 0, 1]])  # x1=1, x2=0, x3=1
y_actual = np.array([[1]])  # Actual output = 1
print("Input:", X)
print("Actual Output:", y_actual)
print("\n" + "="*50)
# Create neural network with different activation functions
# 3 inputs -> 2 hidden neurons (ReLU) -> 1 output (Sigmoid)
model = keras.Sequential([
    layers.Dense(2, activation='relu', input_shape=(3,), name='hidden_layer'),
    layers.Dense(1, activation='sigmoid', name='output_layer')
])
W_hidden = np.array([
    [-0.306, 0.192],   # x1 connections to H4, H5
    [0.4, 0.1],        # x2 connections to H4, H5
    [-0.508, 0.194]    # x3 connections to H4, H5
])
b_hidden = np.array([-0.408, 0.194])  # biases for H4, H5
# Output layer weights
W_output = np.array([[-0.261], [-0.138]])  # H4, H5 to O6
b_output = np.array([0.218])  # bias for O6
# Set weights in the model
model.layers[0].set_weights([W_hidden, b_hidden])
model.layers[1].set_weights([W_output, b_output])
print("\n1. FORWARD PASS (Before Training)")
print("="*50)
# Forward pass to see initial prediction
prediction_before = model.predict(X, verbose=0)
print(f"Predicted Output: {prediction_before[0][0]:.6f}")
print(f"Actual Output: {y_actual[0][0]}")
print(f"Initial Loss: {((y_actual[0][0] - prediction_before[0][0])**2 / 2):.6f}")
# Compile the model
model.compile(
    optimizer=keras.optimizers.SGD(learning_rate=0.5),
    loss='mse',
    metrics=['mae']
)
print("\n2. BACKPROPAGATION (Training)")
print("="*50)
# Train for one epoch to perform backpropagation
history = model.fit(X, y_actual, epochs=1, verbose=1)
print("\n3. FORWARD PASS (After One Backpropagation Step)")
print("="*50)
# Forward pass after training
prediction_after = model.predict(X, verbose=0)
print(f"Predicted Output: {prediction_after[0][0]:.6f}")
print(f"Actual Output: {y_actual[0][0]}")
print(f"Final Loss: {((y_actual[0][0] - prediction_after[0][0])**2 / 2):.6f}")
print("\n4. UPDATED WEIGHTS")
print("="*50)
# Display updated weights
print("\nHidden Layer Weights:")
print(model.layers[0].get_weights()[0])
print("\nHidden Layer Biases:")
print(model.layers[0].get_weights()[1])
print("\nOutput Layer Weights:")
print(model.layers[1].get_weights()[0])
print("\nOutput Layer Biases:")
print(model.layers[1].get_weights()[1])
print("\n" + "="*50)
print("Backpropagation completed!")
print("The weights have been updated to reduce the error.")
```
