```python
import numpy as np
A = np.array([[11, 24, 36],
              [4, 5, 6]])
B = np.array([[10, 20, 30],
              [40, 50, 60]])
C = A + B
print(C)
```

```python
import tensorflow as tf
A = tf.constant([[12,2,3],[45,5,6]])
B=tf.constant([[10,20,30],[40,50,60]])
C=A+B
print(C)
```
