```python

# RNN for Named Entity Recognition (NER)
import torch
import torch.nn as nn
import torch.optim as optim
sentences = [
    ["Sonali", "lives", "in", "Bengaluru"],
    ["Samudra", "lives", "in", "chennai"],
    ["Kani", "lives", "in", "Pune"]
]

tags = [
    ["PER", "O", "O", "LOC"],
    ["PER", "O", "O", "LOC"],
    ["PER", "O", "O", "LOC"]
]
word_to_ix = {}

for sent in sentences:
    for word in sent:
        if word not in word_to_ix:
            word_to_ix[word] = len(word_to_ix)

# Add UNK token
word_to_ix["<UNK>"] = len(word_to_ix)

tag_to_ix = {"PER": 0, "LOC": 1, "ORG": 2, "O": 3}

ix_to_tag = {v: k for k, v in tag_to_ix.items()}

vocab_size = len(word_to_ix)
tagset_size = len(tag_to_ix)
def prepare_sequence(seq, to_ix):
    return torch.tensor(
        [to_ix[w] if w in to_ix else to_ix["<UNK>"] for w in seq],
        dtype=torch.long
    )
# Model
class NER_RNN(nn.Module):
    def __init__(self, vocab_size, tagset_size):
        super(NER_RNN, self).__init__()

        self.embedding = nn.Embedding(vocab_size, 10)
        self.rnn = nn.RNN(10, 16, batch_first=True)
        self.fc = nn.Linear(16, tagset_size)

    def forward(self, sentence):
        embeds = self.embedding(sentence)
        out, _ = self.rnn(embeds)
        tag_space = self.fc(out)
        return tag_space
# Initialize
model = NER_RNN(vocab_size, tagset_size)

loss_function = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=0.1)
# Training
for epoch in range(50):
    total_loss = 0

    for sentence, tag in zip(sentences, tags):

        model.zero_grad()

        sentence_in = prepare_sequence(sentence, word_to_ix)
        targets = prepare_sequence(tag, tag_to_ix)

        outputs = model(sentence_in.unsqueeze(0)).squeeze(0)

        loss = loss_function(outputs, targets)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    print(f"Epoch {epoch+1}, Loss: {total_loss:.4f}")
# Testing
test = ["Sonali", "lives", "in", "Hyderabad"]
test_in = prepare_sequence(test, word_to_ix)
with torch.no_grad():
    outputs = model(test_in.unsqueeze(0)).squeeze(0)
    predicted = torch.argmax(outputs, dim=1)

predicted_tags = [ix_to_tag[i.item()] for i in predicted]

print("\nSentence:", test)
print("Predicted Tags:", predicted_tags)

```
