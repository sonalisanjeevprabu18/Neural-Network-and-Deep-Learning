```python
import tensorflow as tf

# Hidden layer weights (2 neurons)
h1 = tf.constant([0.3, 0.4, -0.1], dtype=tf.float32)
h2 = tf.constant([0.9, 0.7, -0.45], dtype=tf.float32)

# Output layer weights (connects h1 & h2 to final output)
W_out = tf.constant([0.6, -0.3], dtype=tf.float32)
b_out = tf.constant(0.2, dtype=tf.float32)

# Input
x = tf.constant([1.0, 0.0, 1.0], dtype=tf.float32)

# Hidden layer (dot products)
z1 = tf.reduce_sum(x * h1)
z2 = tf.reduce_sum(x * h2)

# Hidden activations (ReLU)
a1 = tf.nn.relu(z1)
a2 = tf.nn.relu(z2)

# Combine hidden neurons â†’ final weighted sum
z_final = a1 * W_out[0] + a2 * W_out[1] + b_out

# Final spam score (Sigmoid)
spam_score = tf.nn.sigmoid(z_final)

# Prediction
prediction = "SPAM" if spam_score.numpy() > 0.5 else "NOT SPAM"

print("Spam Score:", spam_score.numpy())
print("Prediction:", prediction)
```
