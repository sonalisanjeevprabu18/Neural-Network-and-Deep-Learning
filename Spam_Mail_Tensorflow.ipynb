```python
import numpy as np
import tensorflow as tf

W = tf.constant([4.0, 1.5, 3.0])
B = tf.constant(-5.0)

def single_neuron_forward(features):
    weighted_sum = tf.reduce_sum(features * W) + B
    output = tf.keras.activations.relu(weighted_sum)
    return output, weighted_sum

def predict_spam(score):
    return 1 if score.numpy() > 0 else 0

features = tf.constant([1.0, 1.0, 1.0])
output, z = single_neuron_forward(features)
prediction = predict_spam(output)

print(f"Weighted Sum (Z): {z.numpy():.2f}")
print(f"ReLU Output (A):  {output.numpy():.2f}")
print(f"Prediction:     {prediction} ({'SPAM' if prediction else 'NOT SPAM'})\n")
```
